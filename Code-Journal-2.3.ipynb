```python
random_numbers = np.random.uniform(0, 1, 1000)
```

```python
plt.hist(random_numbers, bins=100, edgecolor='black')

plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Uniformly Distributed Random Numbers")

plt.grid(True)
plt.show()
```
